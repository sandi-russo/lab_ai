```python
import random
from typing import Tuple, Dict, Set

Position = Tuple[int, int]


# ===========================
# Ambiente nascosto
# ===========================

def init_hidden_env_2d(rows=5, cols=7, dirt_prob=0.4,
                       agent_start_pos: Position = (0, 0),
                       seed=None):
    if seed is not None:
        random.seed(seed)

    dirt: Dict[Position, bool] = {}
    for r in range(rows):
        for c in range(cols):
            dirt[(r, c)] = (random.random() < dirt_prob)

    agent_pos = agent_start_pos
    performance = 0
    return rows, cols, dirt, agent_pos, performance


def percept_unknown_2d(agent_pos: Position,
                       dirt: Dict[Position, bool],
                       last_action_bumped: bool):
    return agent_pos, dirt[agent_pos], last_action_bumped


def step_hidden_env_2d(rows, cols, dirt,
                       agent_pos, performance, action: str):
    performance -= 1
    bumped = False
    r, c = agent_pos

    if action == "SUCK":
        if dirt[agent_pos]:
            dirt[agent_pos] = False
            performance += 10

    elif action == "LEFT":
        if c > 0:
            c -= 1
        else:
            bumped = True
    elif action == "RIGHT":
        if c < cols - 1:
            c += 1
        else:
            bumped = True
    elif action == "UP":
        if r > 0:
            r -= 1
        else:
            bumped = True
    elif action == "DOWN":
        if r < rows - 1:
            r += 1
        else:
            bumped = True
    elif action == "NOOP":
        pass

    agent_pos = (r, c)
    return dirt, agent_pos, performance, bumped


def all_clean_2d(dirt: Dict[Position, bool]) -> bool:
    return not any(dirt.values())


def count_dirty_2d(dirt: Dict[Position, bool]) -> int:
    return sum(1 for v in dirt.values() if v)


def show_hidden_env_2d(rows, cols, dirt, agent_pos, performance):
    """
    Stampa la griglia con celle a larghezza fissa:
    - " . " per pulito
    - " D " per sporco
    - "(.)" / "(D)" se presente l'agente
    """
    lines = []
    for r in range(rows):
        row_cells = []
        for c in range(cols):
            pos = (r, c)
            dirty_here = dirt[pos]
            if pos == agent_pos:
                ch = "(D)" if dirty_here else "(.)"
            else:
                ch = " D " if dirty_here else " . "
            row_cells.append(ch)
        lines.append("".join(row_cells))
    print("\n".join(lines))
    print(f"(perf={performance}, dirty_left={count_dirty_2d(dirt)})\n")

# ===========================
# Agente minimal con modello
# ===========================

def init_agent_2d_unknown_simple(start_pos: Position):
    """
    Stato minimale:
    - pos: posizione stimata
    - visited: celle visitate
    - walls: celle che l'agente crede essere muri
    - last_action: ultima azione (per interpretare il bump)
    """
    return {
        "pos": start_pos,
        "visited": {start_pos},
        "walls": set(),       # type: Set[Position]
        "last_action": None,
    }


def neighbors4(pos: Position):
    r, c = pos
    return {
        "UP":    (r - 1, c),
        "DOWN":  (r + 1, c),
        "LEFT":  (r, c - 1),
        "RIGHT": (r, c + 1),
    }


def agent_2d_unknown_simple_policy(agent_state, percept):
    """
    Politica semplicissima:
    - se sporco -> SUCK
    - altrimenti prova a muoversi in una direzione che non è nota come muro,
      preferibilmente verso celle mai visitate.
    """
    pos = agent_state["pos"]
    visited: Set[Position] = agent_state["visited"]
    walls: Set[Position] = agent_state["walls"]
    last_action = agent_state["last_action"]

    (obs_pos, is_dirty, bumped) = percept

    # 1. Se c'è stato bump, segna il muro in base a last_action e POS precedente
    if bumped and last_action in {"UP", "DOWN", "LEFT", "RIGHT"}:
        r, c = pos
        if last_action == "UP":
            walls.add((r - 1, c))
        elif last_action == "DOWN":
            walls.add((r + 1, c))
        elif last_action == "LEFT":
            walls.add((r, c - 1))
        elif last_action == "RIGHT":
            walls.add((r, c + 1))

    # 2. Allinea la posizione alla percepita
    pos = obs_pos
    visited.add(pos)

    # 3. Se c'è sporco, aspira
    if is_dirty:
        action = "SUCK"
    else:
        # 4. Scegli una direzione verso una cella adiacente non nota come muro
        neighbors = neighbors4(pos)

        # prima prova celle non visitate e non muro
        candidates = [
            (act, p) for act, p in neighbors.items()
            if p not in walls and p not in visited
        ]
        if not candidates:
            # se tutte visitate, ma non muri, muoviti comunque (evita muri)
            candidates = [
                (act, p) for act, p in neighbors.items()
                if p not in walls
            ]

        if candidates:
            # scegli a caso tra i candidati
            act, _ = random.choice(candidates)
            action = act
        else:
            # circondato da muri (o non ci sono candidati): resta fermo
            action = "NOOP"

    # 5. Aggiorna posizione stimata in base all'azione scelta
    r, c = pos
    new_pos = pos
    if action == "UP":
        new_pos = (r - 1, c)
    elif action == "DOWN":
        new_pos = (r + 1, c)
    elif action == "LEFT":
        new_pos = (r, c - 1)
    elif action == "RIGHT":
        new_pos = (r, c + 1)

    new_state = {
        "pos": new_pos,
        "visited": visited,
        "walls": walls,
        "last_action": action,
    }
    return action, new_state


# ===========================
# Simulazione
# ===========================

def run_episode_2d_unknown_simple(real_rows=5, real_cols=7,
                                  dirt_prob=0.4,
                                  steps=100,
                                  agent_start_pos=(0, 0),
                                  seed=None,
                                  show_real=True):
    rows, cols, dirt, agent_pos, performance = init_hidden_env_2d(
        rows=real_rows,
        cols=real_cols,
        dirt_prob=dirt_prob,
        agent_start_pos=agent_start_pos,
        seed=seed
    )

    if seed is not None:
        random.seed(seed)

    agent_state = init_agent_2d_unknown_simple(start_pos=agent_pos)
    last_bumped = False

    for t in range(steps):
        print(f"t={t:02d}:")
        if show_real:
            print("=== Ambiente vero ===")
            show_hidden_env_2d(rows, cols, dirt, agent_pos, performance)

        percept = percept_unknown_2d(agent_pos, dirt, last_bumped)
        action, agent_state = agent_2d_unknown_simple_policy(agent_state, percept)

        dirt, new_agent_pos, performance, bumped = step_hidden_env_2d(
            rows, cols, dirt, agent_pos, performance, action
        )
        last_bumped = bumped
        agent_pos = new_agent_pos

        print(f"  Azione: {action}, bump={bumped}, pos={agent_pos}\n")

        if all_clean_2d(dirt):
            print(f"t={t+1:02d} (tutto pulito):")
            if show_real:
                show_hidden_env_2d(rows, cols, dirt, agent_pos, performance)
            print("Episodio terminato.")
            break


if __name__ == "__main__":
    run_episode_2d_unknown_simple(
        real_rows=4,
        real_cols=6,
        dirt_prob=0.5,
        steps=80,
        agent_start_pos=(1, 1),
        seed=123,
        show_real=True
    )
```
