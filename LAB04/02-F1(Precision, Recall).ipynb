```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f1(x_val, y_val):
    return 2/(1/x_val + 1/y_val)
```

```python
start = 0.01
stop = 1.01
step = 0.01
x = np.arange(start, stop, step)
y = np.arange(start, stop, step)
#f1 = 2/(1/pr + 1/rc)
Pr, Rc = np.meshgrid(x, y)
Z = f1(Pr, Rc)
```

```python
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')

ax.plot_surface(Pr, Rc, Z)

ax.set_xlabel("Precision")
ax.set_ylabel("Recall")
ax.set_title("F1 score")

plt.show()
```

```python
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
my_cmap = plt.get_cmap('viridis')

ax.plot_surface(Pr, Rc, Z)

surf = ax.plot_surface(Pr, Rc, Z, 
                       rstride = 8,
                       cstride = 8,
                       alpha = 0.8,
                       cmap = my_cmap)
cset = ax.contourf(Pr, Rc, Z,
                   zdir ='z',
                   offset = np.min(Z),
                   cmap = my_cmap)
cset = ax.contourf(Pr, Rc, Z,
                   zdir ='x',
                   offset = np.min(Pr),
                   cmap = my_cmap)
cset = ax.contourf(Pr, Rc, Z, 
                   zdir ='y',
                   offset =  np.max(Rc),
                   cmap = my_cmap)
fig.colorbar(surf, ax = ax, 
             shrink = 0.5,
             aspect = 5)

ax.set_xlabel("Precision")
ax.set_ylabel("Recall")
ax.set_title("F1 score")

plt.show()
```
